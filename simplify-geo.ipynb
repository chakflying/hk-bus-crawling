```python
import geopandas

gdf = geopandas.read_file("1380-1.json")

gdf.set_crs("EPSG:4326")
```

```python
gdf.geometry.get_coordinates()
```

```python
simplified = gdf.simplify(tolerance=0.000001)
simplified.get_coordinates()
```

```python
from shapely import transform

rounded = transform(simplified, lambda x: x.round(5))
```

```python
updated = gdf.set_geometry(rounded)
updated.to_file("1380-1-simplified.geojson", driver="GeoJSON")
```
